# **Bonus Task 7 – RFM (Recency, Frequency, Monetary) Customer Segmentation**

**Step 7.1 Find the Latest Order Date**

Use the latest order date in the dataset as the reference date for calculating recency.

```python
# ============================================================
# RFM Analysis
# ============================================================

reference_date = orders_df['Order Date'].max() + pd.Timedelta(days=1)

print("Reference Date:", reference_date)
```

**Step 7.2 Create the RFM Table**

```python
rfm = orders_df.groupby('Customer ID').agg({
    'Order Date': lambda x: (reference_date - x.max()).days,
    'Order ID': 'nunique',
    'Sales': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

rfm.head()
```

**Step 7.3 Summary Statistics**

```python
rfm.describe()
```

**Step 7.4 Calculate RFM Scores**

We'll divide each metric into four groups using quartiles.

```python
# Recency Score (Lower is Better)
rfm['R_Score'] = pd.qcut(
    rfm['Recency'],
    4,
    labels=[4, 3, 2, 1]
).astype(int)

# Frequency Score (Higher is Better)
rfm['F_Score'] = pd.qcut(
    rfm['Frequency'].rank(method='first'),
    4,
    labels=[1, 2, 3, 4]
).astype(int)

# Monetary Score (Higher is Better)
rfm['M_Score'] = pd.qcut(
    rfm['Monetary'],
    4,
    labels=[1, 2, 3, 4]
).astype(int)
```

**Step 7.5 Create the Overall RFM Score**

```python
rfm['RFM Score'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)

rfm.head()
```

**Step 7.6 Customer Segmentation**

We'll classify customers based on their RFM scores.

```python
def customer_segment(row):

    if row['RFM Score'] == '444':
        return 'Champions'

    elif row['R_Score'] >= 3 and row['F_Score'] >= 3:
        return 'Loyal Customers'

    elif row['R_Score'] == 4:
        return 'Recent Customers'

    elif row['M_Score'] == 4:
        return 'Big Spenders'

    else:
        return 'Others'


rfm['Customer Segment'] = rfm.apply(customer_segment, axis=1)
```

**Step 7.7 Number of Customers in Each Segment**

Count how many customers belong to each segment.

```python
# Count customers in each segment

segment_counts = rfm['Customer Segment'].value_counts()

segment_counts
```

# **Bonus Task 8 – Profit Prediction using Linear Regression**

**Step 8.1 Select Features**

```python
# ============================================================
# Feature Selection
# ============================================================

ml_df = orders_df[[
    'Sales',
    'Quantity',
    'Discount',
    'Category',
    'Sub-Category',
    'Region',
    'Segment',
    'Ship Mode',
    'Profit'
]]

ml_df.head()
```

**Step 8.2 Check Missing Values**

```python
ml_df.isnull().sum()
```

**Step 8.3 Encode Categorical Variables**

Machine Learning algorithms require numeric input.

We'll use One-Hot Encoding.

```python
# Convert categorical columns into dummy variables

ml_df = pd.get_dummies(
    ml_df,
    columns=[
        'Category',
        'Sub-Category',
        'Region',
        'Segment',
        'Ship Mode'
    ],
    drop_first=True
)

ml_df.head()
```

**Step 8.4 Split Features and Target**

```python
# Features

X = ml_df.drop('Profit', axis=1)

# Target

y = ml_df['Profit']
```

**Step 8.5 Train-Test Split**

We'll use:
*  80% Training
*  20% Testing

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Samples :", X_train.shape[0])
print("Testing Samples  :", X_test.shape[0])
```

**Step 8.6 Train Linear Regression Model**

```python
# Train Model

lr_model = LinearRegression()

lr_model.fit(
    X_train,
    y_train
)

print("Linear Regression Model Trained Successfully!")
```

**Step 8.7 Predict Profit**

```python
# Predict

y_pred = lr_model.predict(X_test)

y_pred[:10]
```

**Step 8.8 Evaluate Model**

Calculate:
*  MAE
*  RMSE
*  R² Score

```python
# Mean Absolute Error

mae = mean_absolute_error(
    y_test,
    y_pred
)

# Root Mean Squared Error

rmse = np.sqrt(
    mean_squared_error(
        y_test,
        y_pred
    )
)

# R² Score

r2 = r2_score(
    y_test,
    y_pred
)

print(f"Mean Absolute Error : {mae:.2f}")
print(f"Root Mean Squared Error : {rmse:.2f}")
print(f"R² Score : {r2:.4f}")
```

**Step 8.9 Actual vs Predicted Values**

```python
comparison = pd.DataFrame({
    'Actual Profit': y_test.values,
    'Predicted Profit': y_pred
})

comparison.head(10)
```

**Step 8.10 Scatter Plot**

This is one of the most commonly used plots in regression projects.

```python
plt.figure(figsize=(8,6))

plt.scatter(
    y_test,
    y_pred,
    alpha=0.6
)

plt.xlabel("Actual Profit")
plt.ylabel("Predicted Profit")

plt.title(
    "Actual vs Predicted Profit",
    fontsize=15,
    fontweight='bold'
)

plt.grid(True)

plt.tight_layout()
plt.savefig("Actual_vs_Predicted_Profit.png", dpi=300, bbox_inches="tight")

plt.show()
```

**Step 8.11 Feature Importance (Coefficients)**

Linear Regression provides coefficients that indicate how each feature influences the predicted profit.

```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lr_model.coef_
})

coefficients = coefficients.sort_values(
    by='Coefficient',
    ascending=False
)

coefficients.head(15)
```

**Step 8.12 Visualize Top Feature Coefficients**

```python
top_features = coefficients.head(15)

plt.figure(figsize=(10,6))

plt.barh(
    top_features['Feature'],
    top_features['Coefficient'],
    edgecolor='black'
)

plt.title(
    "Top Positive Feature Coefficients",
    fontsize=15,
    fontweight='bold'
)

plt.xlabel("Coefficient")

plt.tight_layout()
plt.savefig("Top_Feature_Coefficients.png", dpi=300, bbox_inches="tight")

plt.show()
```
